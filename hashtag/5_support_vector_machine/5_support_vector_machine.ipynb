# O SVM (Support Vector Machine)
- Utilizando o dataset iris, podemos utilizar apenas a coluna petal width (cm) e visualizar a seguinte classificao dos dados:

<br>
- Se tivermos um novo ponto, como o mostrado em vermelho, automaticamente j vamos pensar que esse ponto possui a mesma classe dos pontos em roxo:

<br>
- Da mesma forma, tendemos a pensar que esse novo ponto deve ser classificado como amarelo:

<br>
- E  uma lgica muito parecida com essa que temos no SVM
<br>
- Podemos basicamente pegar os pontos que esto mais no extremo dos dados (**os pontos de duas classes diferentes que esto mais prximos entre si**) e usar esses pontos para determinar um <font color='red'>**hiperplano**</font> capaz de separar essas classes

- **Quanto maior a margem, melhor!**
<br><br>
- Um fato importante  que ele **pode ser utilizado tanto em dados linearmente separveis** (como mostrado acima) **como em dados que no so linearmente separveis** (que so a maioria dos nossos dados)
    - Ele faz isso elevando as dimenses dos dados
    - O SVM vai usar as funes de kernel e no prprio algoritmo ele vai buscar a melhor dimenso capaz de encontrar essa reta
    - S que ele no faz efetivamente essas transformaes, ele s vai calcular a relao entre os pontos considerando que eles esto nessas dimenses. Isso  chamado de **"Truque do Kernel"**
<br>

<br>
- E por mais que a gente tenha visto em apenas 1D, ele funciona para qualquer dimenso dos dados

<br>
- ** um algoritmo muito poderoso porm pode demorar muito tempo para treinar os dados**

- **Para comear, podemos j usar o dataset iris**
    - https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris

```python
# Importando o dataset
from sklearn.datasets import load_iris
X,y = load_iris(return_X_y = True, as_frame=True)
```

```python
# Considerando as duas colunas de ptala os targets 0 e 1
X = X.loc[y.isin([0,1]),['petal length (cm)','petal width (cm)']]
y = y[y.isin([0,1])]
```

- E separar em treino e teste
    - https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

```python
# Separando em treino e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

- Visualizando graficamente essas informaes
    - https://matplotlib.org/stable/plot_types/basic/scatter_plot.html#sphx-glr-plot-types-basic-scatter-plot-py

```python
# Visualizando graficamente
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

ax.scatter(X_train['petal length (cm)'], X_train['petal width (cm)'],c=y_train,s=60)

plt.show()
```

- **Utilizando o SVC**
    - https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC

```python
# Importando o SVC
from sklearn.svm import SVC
```

```python
# Criando o classificador e j fazendo o fit com os dados
clf = SVC(kernel='linear').fit(X_train,y_train)
```

```python
# Visualizando o coef
clf.coef_[0]
```

```python
# Separando w1
w1 = clf.coef_[0][0]
```

```python
# e w2
w2 = clf.coef_[0][1]
```

```python
# E agora utilizando o intercept_ como w0
w0 = clf.intercept_[0]
```

- Para traar a reta, podemos considerar que a equao da reta ser dada por:
    - y = ax + b
- Para considerar os dois coeficientes, tambm podemos escrever como:
    - w1.X_train[0] + w2.X_train[1] + w0 = 0
    - w1.X_train['petal width (cm)'] + w2.X_train['petal length (cm)'] + w0 = 0
    - Sendo x = X_train['petal width (cm)'] e y = X_train['petal length (cm)']:
        - w1.x + w2.y + w0 = 0
        - w2.y = -w1.x - w0
        - y = (-w1.x - w0)/w2

```python
# Traando essa reta no grfico
fig, ax = plt.subplots()

ax.scatter(X_train['petal length (cm)'], X_train['petal width (cm)'],c=y_train,s=60)

x = np.linspace(1,5,100)
y = (-w1*x-w0)/w2
ax.plot(x,y,'r')

ax.set(ylim=(0,1.7))

plt.show()
```

```python
# Alm disso, tambm podemos ver quais foram os vetores de suporte utilizados para traar essa reta
clf.support_vectors_
```

```python
# Pegando os valores de x
clf.support_vectors_[:,0]
```

```python
# E os valores de y
clf.support_vectors_[:,1]
```

```python
# Visualizando esses pontos
fig, ax = plt.subplots()

ax.scatter(X_train['petal length (cm)'], X_train['petal width (cm)'],c=y_train,s=60)

x = np.linspace(1,5,100)
y = (-w1*x-w0)/w2
ax.plot(x,y,'r')
y2 = (+1-w1*x-w0)/w2
ax.plot(x,y2,'--r')
y3 = (-1-w1*x-w0)/w2
ax.plot(x,y3,'--r')

ax.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],c='r')

ax.set(ylim=(0,1.7))

plt.show()
```

- **Agora podemos usar toda a nossa base de dados**

```python
# Importando o dataset
X,y = load_iris(return_X_y = True, as_frame=True)
```

```python
# Separando em treino e teste
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
# Criando o classificador
clf2 = SVC().fit(X_train,y_train)
```

```python
# Fazendo a previso
y_pred = clf2.predict(X_test)
```

```python
# E analisando a matriz de confuso
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
